##### Testing Accuracy of Shapiro - Wilks Test on Randomly Generated Data
Shapiro Wilks Test tests whether a data sample follows a gauissian distribution (normal distribution)

https://www.analyticsvidhya.com/blog/2020/04/how-to-generate-random-numbers-in-python/

```python
#import required libraries
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Randomly generate three lists of data
#Create empty lists
randomlist1 = []
randomlist2 = []
randomlist3 = []
#for loop to append random numbers to empty lists
for i in range(0,100):
    x = random.randint(1,5000)
    randomlist1.append(x)
for i in range(0,100):
    y = random.randint(1,10000)
    randomlist2.append(y)
for i in range(0,100):
    z = random.randint(1,20000)
    randomlist3.append(z)
#Create an array of normaly distributed randome numbers
gauissian = np.random.normal(loc = 0, scale = 1, size = 100)
#Convert np.array to a list
random_normal = gauissian.tolist()

#Convert Lists into a Dictionary
Dictionary = {'X1': randomlist1, 'Y1': randomlist2, 'X2': randomlist3, 'Y2': random_normal }
#Convert Dictionary to pandas dataframe
df = pd.DataFrame(Dictionary)
print(df, df.info())
```

###### Visualizing DIstributions

```python
#Kernel density plot
sns.distplot(df['X1'], kde = True, hist = False, color = 'b')
sns.set_style('darkgrid')
display(plt.show(), df['X1'].describe())
```

```python
#Kernel density plot
sns.distplot(df['Y1'], kde = True, hist = False, color = 'b')
sns.set_style('darkgrid')
display(plt.show(), df['Y1'].describe())
```

```python
#Kernel density plot
sns.distplot(df['X2'], kde = True, hist = False, color = 'b')
sns.set_style('darkgrid')
display(plt.show(), df['X2'].describe())
```

```python
#Kernel density plot
sns.distplot(df['Y2'], kde = True, hist = False, color = 'r')
sns.set_style('darkgrid')
display(plt.show(), df['Y2'].describe())
```

As we can see from the above kernel density plot visualizations three out four features follow a gaussian distribution.

##### Shapiro-Wilk Test
This statistical test will test whether a data sample has a Gaussian distribution.

Assumptions:

Observations in each sample are independent and identically distributed

Interpretation:

H0: the sample has a Gaussian distribution.

H1: the sample does not have a Gaussian distribution.

```python
#Interpretation
#H0: the sample has a Gaussian distribution.
#H1: the sample does not have a Gaussian distribution.
from scipy.stats import shapiro
test_1 = df['X1']
stat, p = shapiro(test_1)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('X1 variable probable to follow a Gaussian Distribution')
else:
    print('X1 variable probable not to follow a Gaussian Distribution')
```

```python
from scipy.stats import shapiro
test_1 = df['Y1']
stat, p = shapiro(test_1)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Y1 variable probable to follow a Gaussian Distribution')
else:
    print('Y1 variable probable not to follow a Gaussian Distribution')
```

```python
from scipy.stats import shapiro
test_1 = df['X2']
stat, p = shapiro(test_1)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('X2 variable probable to follow a Gaussian Distribution')
else:
    print('X2 variable probable not to follow a Gaussian Distribution')
```

```python
from scipy.stats import shapiro
test_1 = df['Y2']
stat, p = shapiro(test_1)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Y2 variable probable to follow a Gaussian Distribution')
else:
    print('Y2 variable probable not to follow a Gaussian Distribution')
```

We can confirm that the testing envoirnment has been succesful in determining accurately whether individual data features follow a gaussian distribution.
