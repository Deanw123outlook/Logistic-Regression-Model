#### Encoding Testing and Analysis

```python
#import libraries
import numpy as np
import pandas as pd
```

```python
#Import csv file as pandas dataframe
data = pd.read_csv('healthcare-dataset-stroke-data.csv')
#Create copy of Original DataFrame
data_original = data.copy(deep=True)
#data type conversions
data['hypertension'] = data['hypertension'].astype('object')
data['stroke'] = data['stroke'].astype('object')
#print head of dataset
data.head()
```

```python
data.describe()
```

```python
data.describe(include = object)
```

##### One Hot Encoding

```python
#One Hot Encoding
one_hot_encoded = pd.get_dummies(data)
one_hot_encoded
```

The original data has increased from 12 columns to 25 columns after one hot encoding. Next we will take a look at dummy encoding.

##### Dummy Encoding

```python
#dummy encoding
dummy_encoded_data = pd.get_dummies(data, drop_first = True)
dummy_encoded_data
```

The original data has increased from 12 columns to 18 columns after dummy encoding.
